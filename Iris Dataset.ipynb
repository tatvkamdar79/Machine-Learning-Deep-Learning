```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd
import seaborn as sbn

dataset_iris = load_iris() # loading dataset
df = pd.DataFrame(np.c_[dataset_iris['data'], dataset_iris['target']], columns=dataset_iris.feature_names + ['target'])
```

```python
print("\n Keys of Iris Dataset: \n {}".format(dataset_iris.keys()))
print("\n Features: \n",dataset_iris['feature_names'])
print("\n First 5 Columns of the data: \n", dataset_iris['data'][:5])
```

```python
X_train, X_test, Y_train, Y_test = \
        train_test_split(dataset_iris['data'], dataset_iris['target'], 
                        random_state=33, shuffle=True, test_size=0.2)
```

```python
sbn.pairplot(df, hue='target', height = 3.0)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, Y_train)
print("Test set score: {:.2f}".format(knn.score(X_test, Y_test)))
```
